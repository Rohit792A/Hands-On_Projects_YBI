##**Hand Written Digit Prediction - Classifiication Analysis**


**Attributes Information**

The didgit dataset consists of 8x8 pixel images of digits .the images attribute of the dataset stores 8x8 arrays of grayscale values for each image. We will use these arrays to visualize the first 4 images . The target attribute of the dataset stores the digit each image represents.


##**Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##**Import Data**

```python
from sklearn.datasets import load_digits
```

```python
df = load_digits()
```

## **Data Preproccessing**

**Flatten Image**

```python
df.images.shape
```

```python
df.images[0]
```

```python
df.images[0].shape
```

```python
n_samples = len(df.images)
data = df.images.reshape(n_samples,-1)
```

```python
data[0]
```

```python
data[0].shape
```

```python
data.shape
```

## **Scaling the Data**

```python
data.min()
```

```python
data.max()
```

```python
data=data/16
```

```python
data.min()
```

```python
data.max()
```

```python
data[0]
```

##**Train Test Split Data**

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train , X_test, y_train , y_test =train_test_split(data ,df.target , train_size=0.7,random_state=182529)
```

```python
X_train.shape , X_test.shape, y_train.shape , y_test.shape
```

## **Random  Forest Model**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

##**Predict Test Data**

```python
y_pred=rf.predict(X_test)
```

```python
y_pred
```


##**Model Accuracy**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
